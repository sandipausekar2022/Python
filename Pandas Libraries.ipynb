# Pandas 
1. Pandas is an open-source library that is made mainly for working with relational or labeled data both easily and intuitively.
2. It provides various data structures and operations for manipulating numerical data and time series.
3. This library is built on top of the NumPy library.
4. Pandas is fast and it has high performance & productivity for users.
5. Pandas generally provide two data structures for manipulating data, They are:
 * Series
 * DataFrame

# Series 
 * Series is a one-dimensional labeled array and capable of holding data of any type (integer, string, float, python objects, etc.)

```python
# Series creation
```

```python
lst=[1,2,3,4,5]
type(lst)
```

```python
s=pd.Series(lst)
type(s)
```

```python
s
```

```python
s=pd.Series(data=['A',10,2500])
s
```

```python
d={1:'A',2:'B',3:'C'}
type(d)
```

```python
s=pd.Series(d)
s
```

```python
s=pd.Series(data=[10,20,30],index=['A','B','C'],name='My series')
s
```

```python
# Accessing elements from a series using index labels.
```

```python
s[0]
```

```python
s['C']
```

# DataFrame
 * A Pandas DataFrame is a 2 dimensional data structure, like a 2 dimensional array, or a table with rows and columns.

# Ways to create pandas Dataframe
1. By creating a nested lists and colname lists.
2. By creating a dictionary

```python
lst=[['pyton',90,'pune'],['ML',100,'Mumbai'],['DL',95,'Hyd']]

col=['Subject','Marks','City']
```

```python
df = pd.DataFrame(data=lst, columns=col)
```
